```python
import nltk

# Ensure the required resources are downloaded
nltk.download('punkt')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
```

```python
text = "4780 Stillwell Beckett Rd, Oxford OH is not a valid address. Also, Maps takes people, literally, to a tree when they try to get to my house. and in Gujrat Baroda got raid by funky peple .. near Ajaawa we got about thisAbout 1/2 the time when I put in my address (similar to this fake one but a few miles away) maps will change to using this fake one. This really needs to be fixed but because I can't talk with someone over the phone it's difficult to explain - but it is 1000% wrong - there is no such address as 4780 Stillwell Beckett Rd, Oxford OH even though it shows up on Google Maps."
text
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
tagged_sentences = [nltk.pos_tag(sentence) for sentence in tokenized_sentences]
```

```python

chunked_sentences = [nltk.ne_chunk(tagged) for tagged in tagged_sentences]
```

```python
named_entities = []

for chunked_sentence in chunked_sentences:
    for chunk in chunked_sentence:
        if hasattr(chunk, 'label') and chunk.label() in ['GPE', 'LOCATION', 'ORGANIZATION']:
            named_entity = ' '.join(c[0] for c in chunk)
            named_entities.append((named_entity, chunk.label()))

# Print the extracted named entities
print(named_entities)
```
