```python
total_sales_country_customer = df.groupby(['Country', 'CustomerID'])['total sales'].sum().abs()
print(total_sales_country_customer)
```

```python
mean_of_sales = df.groupby(['Country'])['total sales'].mean()
print(mean_of_sales)
```

```python
std_dev = df.groupby(['Country'])['total sales'].std()
print(std_dev)
```

```python
### baki che karvano

import matplotlib.pyplot as plt

# Extract data from the DataFrame
countries = mean_of_sales.index.to_list()
mean_values = mean_of_sales.values
std_dev_values = std_dev.values

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create the first bar chart for mean values
ax.bar(countries, mean_values, label="Mean Sales", color="skyblue", alpha=1.0)

# Create the second bar chart for standard deviation values
ax.bar(countries, std_dev_values, label="Standard Deviation", color="coral", alpha=0.5)

# Add labels and title
ax.set_xlabel("Countries")
ax.set_ylabel("Sales")
ax.set_title("Comparison of Mean Sales and Standard Deviation by Country")

# Add legend and show the plot
ax.legend()
plt.xticks(rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('/content/sample_data/data - Copy.csv' , encoding='latin1')
```

```python



df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Convert the datetime format to dd-mm-yy
df['InvoiceDate'] = df['InvoiceDate'].dt.strftime('%d-%m-%y')


```

```python
print(df['InvoiceDate'])
```

Custom Time And Date records

```python

# Convert 'InvoiceDate' to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

start_date = '12-01-2010'
end_date = '12-02-2010'


# Filter the DataFrame based on the custom range
custom_range_data = df[(df['InvoiceDate'] >= start_date ) &
                       (df['InvoiceDate'] <= end_date)]

print(custom_range_data)
```

```python

# Convert 'InvoiceDate' to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

start_date = '12-01-2010'
end_date = '12-02-2010'
start_time = '08:26:00'
end_time = '9:00:00'

# Filter the DataFrame based on the custom range
customrange_dt = df[(df['InvoiceDate'] >= start_date + ' ' + start_time) &
                       (df['InvoiceDate'] <= end_date + ' ' + end_time)]

print(customrange_dt)
```

```python

print(custom_range_data['total sales'].mean())
print(custom_range_data['UnitPrice'].count())
```

```python


custom_range_data_by_country = custom_range_data.groupby('Country')
print(custom_range_data_by_country['total sales'].mean())
```

```python
plt.plot(custom_range_data_by_country['total sales'].mean())


plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.title('Custom Range Data by Country')
plt.xticks(rotation=90)

plt.show()
```

```python
country_mean = df.groupby('Country')['total sales'].mean()
print(country_mean)
```

```python

import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(country_mean.index, country_mean.values, label="Country Mean", color="skyblue", alpha=1.0)

ax.plot(custom_range_data_by_country['total sales'].mean().index, custom_range_data_by_country['total sales'].mean().values, label="Custom Range Mean", color="coral", alpha=0.5)


ax.set_xlabel("Country")
ax.set_ylabel("Total Sales")
ax.set_title("Comparison of Country Mean and Custom Range Mean")

# Add legend and show the plot
ax.legend()
plt.xticks(rotation=90, ha="right")
plt.tight_layout()
plt.show()
```
